MISSING VALUES

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)        # Numeric column
df['Email'].fillna('unknown@example.com', inplace=True) # String column
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)  # Most common gender
```

```python
df.fillna(method='ffill', inplace=True)  # Fill using previous row
df.fillna(method='bfill', inplace=True)  # Fill using next row
```

```python
df.fillna(method='ffill', inplace=True)  # Fill using previous row
df.fillna(method='bfill', inplace=True)  # Fill using next row
```

```python
df.dropna(inplace=True)           # Drop all rows with any NaN
df.dropna(subset=['Join Date'])   # Drop rows where Join Date is missing
```

Remove duplicate rows

```python
df.duplicated().sum()
```

```python
print("Duplicates after removal:", df.duplicated().sum())
```

Standardize text values

```python
df['Gender'] = df['Gender'].str.strip().str.lower()
df['Gender'] = df['Gender'].replace({
    'm': 'male',
    'male': 'male',
    'f': 'female',
    'female': 'female',
    'fem': 'female',
    'man': 'male',
    'woman': 'female'
})

# Optional: Capitalize for neatness (Male, Female)
df['Gender'] = df['Gender'].str.capitalize()
```

```python
df['Country'] = df['Country'].str.strip().str.lower()

# Replace inconsistent country names with standard ones
df['Country'] = df['Country'].replace({
    'usa': 'united states',
    'u.s.': 'united states',
    'united states': 'united states',
    'uk': 'united kingdom',
    'u.k.': 'united kingdom',
    'ind': 'india',
    'bharat': 'india'
})

# Optional: Capitalize country names (India, United States, etc.)
df['Country'] = df['Country'].apply(lambda x: ' '.join(word.capitalize() for word in x.split()))
```

```python
print("Unique Genders:", df['Gender'].unique())
print("Unique Countries:", df['Country'].unique())
```

CONVERT DATE FORMATE INTO DD-MM-YY

```python
df['Join Date'] = pd.to_datetime(df['Join Date'], errors='coerce')  # handles mixed formats & nulls
df['Join Date'] = df['Join Date'].dt.strftime('%d-%m-%y')
```

Rename column headers to be clean and uniform (e.g., lowercase, no spaces).


```python
df = pd.read_excel("raw_dataset_1.xlsx")  # or read_csv if it's a CSV
print(df.columns)
df.columns = (
    df.columns
    .str.strip()                              # Remove leading/trailing spaces
    .str.upper()                              # Convert to UPPERCASE
    .str.replace(' ', '_')                    # Replace spaces with underscores
    .str.replace('[^A-Z0-9_]', '', regex=True)  # Remove special characters
)
print(df.columns)
```
